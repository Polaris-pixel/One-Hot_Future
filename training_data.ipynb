```python
emissions = emissions.dropna()
```

```python
us_df = emissions.iloc[[2]]
us_df
```

```python
us_df_t=us_df.T
us_df_t
```

```python
us_df_t.reset_index(level=0, inplace=True)
us_df_t
```

```python
us_df_t.rename(columns={'index':'year', 2:'emissions'})
```

```python
us_df_t = us_df_t.iloc[1:]
us_df_t
```

```python
us_df_new = us_df_t.rename(columns={'index':'year', 2:'emissions'})
us_df_new
```

```python
col = us_df_new['year'] 
col_split =[x.split('.') for x in col]
split_df = pd.DataFrame(col_split)
# us_df_new[['years','extra']] = us_df_new[[x.split('.') for x in (us_df_new['year'])]]
split_df
```

```python
split_df.iloc[:,0]
```

```python
split_df.index += 1 
```

```python
us_df_new[['years']]=split_df.iloc[:,0]
us_df_new
```

```python
us_df_new['years'] = pd.to_numeric(us_df_new['years'])
us_df_new
```

```python
# us_df_new.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

X = us_df_new.years.values.reshape(-1, 1)
y = us_df_new.emissions.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
X
```

```python
# Plot the data
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(X, y)
print(model)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_) 
```

```python
predictions = model.predict(X)
print(f"True output: {y[0]}")
print(f"Predicted output: {predictions[0]}")
print(f"Prediction Error: {predictions[0]-y[0]}")
```

```python
x_min = X.min()
x_max = X.max()
```

```python
y_min_actual = y.min()
y_max_actual = y.max()
```

```python
y_min = -49529.98836056 + 27.36157673 * x_min
y_max = -49529.98836056 + 27.36157673 * x_max

print(f"Actual Min Value: {y_min_actual}")
print(f"Calculated Min Value: {round(y_min, 8)}")
print(f"Actual Max Value: {y_max_actual}")
print(f"Calculated Max Value: {y_max}")
```

```python
y_min_predicted = model.predict([[x_min]])
y_max_predicted = model.predict([[x_max]])
print(f"Actual Min Value: {y_min_actual}")
print(f"Predicted Min Value: {y_min_predicted}")
print(f"Actual Max Value: {y_max_actual}")
print(f"Predicted Max Value: {y_max_predicted}")
```
