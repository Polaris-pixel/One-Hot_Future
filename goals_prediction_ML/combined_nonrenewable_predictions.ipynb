```python
import pandas as pd
from functools import reduce
```

```python
nonrenew = pd.read_csv('../clean_data/nonrenewable_energy.csv')
nonrenew
```

```python
nonrenew.at[102, "country"] = "EU"
```

```python
nonrenew.at[99, "country"] = "Global"
```

```python
nonrenew
```

```python
nonrenewable_eu = pd.read_csv('predictions_data_files/nonrenewable_eu.csv', index_col=0)
nonrenewable_eu
```

```python
nonrenewable_global = pd.read_csv('predictions_data_files/nonrenewable_global.csv', index_col=0)
nonrenewable_china = pd.read_csv('predictions_data_files/nonrenewable_china.csv', index_col=0)
nonrenewable_india = pd.read_csv('predictions_data_files/nonrenewable_india.csv', index_col=0)
nonrenewable_us = pd.read_csv('predictions_data_files/nonrenewable_us.csv', index_col=0)
```

```python
data_frames = [nonrenewable_eu, nonrenewable_global, nonrenewable_china, nonrenewable_india, nonrenewable_us]
nonrenew_combined = reduce(lambda  left,right: pd.merge(left,right,on=['years'],
                                            how='outer'), data_frames)
nonrenew_combined
```

```python
nonrenew_combined_T = nonrenew_combined.T
nonrenew_combined_T
```

```python
nonrenew_combined_T= nonrenew_combined_T.iloc[1:]
nonrenew_combined_T
```

```python
nonrenew_combined_T.reset_index(level=0, inplace=True)
```

```python
nonrenew_combined_T.columns = ['country','2020_nonrenewable_predictions', '2030_nonrenewable_predictions', '2040_nonrenewable_predictions', '2050_nonrenewable_predictions']
nonrenew_combined_T
```

```python
nonrenew
```

```python
nonrenewable__predictions= pd.merge(nonrenew, nonrenew_combined_T, on='country' )
nonrenewable__predictions
```

```python
nonrenewable__predictions.to_csv('nonrenewable__predictions.csv', encoding='utf-8')
```
