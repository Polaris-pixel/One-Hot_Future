```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
temp = pd.read_csv("cleaned_temp.csv")
```

```python
ja_data = temp[['Year', 'Japan']]
ja_data = ja_data.dropna()
```

```python
X = ja_data.Year.values.reshape(-1, 1)
y = ja_data.Japan.values.reshape(-1, 1)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
# Fit the model to the data
model.fit(X, y)
```

```python
x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
```

```python
# Calculate the y_min and y_max using model.predict and x_min and x_max
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
# Plot X and y using plt.scatter
# Plot the model fit line using [x_min[0], x_max[0]], [y_min[0], y_max[0]]
plt.scatter(X, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
```

```python
y_2030 = model.predict([[2030]])
y_2040 = model.predict([[2040]])
y_2050 = model.predict([[2050]])
```

```python
my_array1 = y_2030
my_array2 = y_2040
my_array3 = y_2050

df1 = pd.DataFrame(my_array1, columns = ['2030'])
df2 = pd.DataFrame(my_array2, columns = ['2040'])
df3 = pd.DataFrame(my_array3, columns = ['2050'])

frames = [df1, df2, df3]
result = pd.concat(frames, axis=1).T
reset = result.reset_index()
# You need to rename the 0 with the same name as the column in the country df
renamed = reset.rename(columns={"index": "Year", 0: "Japan"})

# renamed.head()
# eu_data.head()

# Create new df
data = [ja_data, renamed]
all_data = pd.concat(data, axis=0)
all_data
```

```python
# Save .csv
ja_output = all_data.to_csv("ja_temp.csv")
```
