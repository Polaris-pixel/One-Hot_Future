```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame
energy_breakdown = pd.read_csv('clean_data/renewable_nonrenewable_breakdown_emissions.csv')
energy_breakdown
```

```python
list(energy_breakdown.columns)
```

```python
df1 = energy_breakdown.iloc[[99]]
df1
```

```python
df1 = df1.filter(regex='hydroelec$',axis=1)
df1.head()
```

```python
df1_t =df1.T
df1_t
```

```python
df1_t.reset_index(level=0, inplace=True)
df1_t
```

```python
df1_t.rename(columns={'index':'year', 99:'hydroelec'})
```

```python
df1_t = df1_t.iloc[1:]
df1_t
```

```python
df1_new = df1_t.rename(columns={'index':'year', 99:'hydroelec'})
df1_new
```

```python
col1 = df1_new['year'] 
col_split1 =[x.split('.') for x in col1]
split_df1 = pd.DataFrame(col_split1)
# us_df_new[['years','extra']] = us_df_new[[x.split('.') for x in (us_df_new['year'])]]
split_df1
```

```python
split_df1.iloc[:,0]
```

```python
split_df1.index += 1 
```

```python
df1_new[['years']]=split_df1.iloc[:,0]
df1_new
```

```python
df1_new['years'] = df1_new['years'].astype(int)
df1_new
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

X = df1_new.years.values.reshape(-1, 1)
y = df1_new.hydroelec.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
X
```

```python
# Plot the data
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(X, y)
print(model)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_) 
```

```python
predictions = model.predict(X)
print(f"True output: {y[0]}")
print(f"Predicted output: {predictions[0]}")
print(f"Prediction Error: {predictions[0]-y[0]}")
```

```python
x_min = X.min()
x_max = X.max()
```

```python
y_min_actual = y.min()
y_max_actual = y.max()
```

```python
y_min = -994.80558188 + 0.51088612 * x_min
y_max = -994.80558188 + 0.51088612 * x_max

print(f"Actual Min Value: {y_min_actual}")
print(f"Calculated Min Value: {round(y_min, 8)}")
print(f"Actual Max Value: {y_max_actual}")
print(f"Calculated Max Value: {y_max}")
```

```python
y_min_predicted = model.predict([[x_min]])
y_max_predicted = model.predict([[x_max]])
print(f"Actual Min Value: {y_min_actual}")
print(f"Predicted Min Value: {y_min_predicted}")
print(f"Actual Max Value: {y_max_actual}")
print(f"Predicted Max Value: {y_max_predicted}")
```

```python
plt.scatter(X, y, c='blue')
plt.plot([x_min, x_max], [y_min, y_max], c='red')
```
