```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read csv
emissions_df = pd.read_csv("clean_data/bp_all.csv")
emissions_df
```

```python
emissions_df_world = emissions_df.iloc[[99]]
emissions_df_world
```

```python
# Pull out data by year prefix
year1965 = emissions_df_world.filter(regex='_all_energy$',axis=1).head()
year1965T = year1965.T
year1965T = year1965T.reset_index(drop=False)
```

```python
year1965T = year1965T.rename(columns={"index":"year", 99: "energy_consumption"})
year1965T
```

```python
col = year1965T['year'] 
col_split =[x.split('.') for x in col]
split_df = pd.DataFrame(col_split)
```

```python
split_df.iloc[:,0]
```

```python
year1965T[['years']]=split_df.iloc[:,0]
year1965T
```

```python
year1965T[['years']]=pd.to_numeric(year1965T['years'])
```

```python
X = year1965T.years.values.reshape(-1, 1)
y = year1965T.energy_consumption.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
X
```

```python
plt.scatter(X, y)
```

```python
# Create the model and fit the model to the data
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Fit the model to the data. 
# Note: This is the training step where you fit the line to the data.
model.fit(X, y)
```

```python
# Print the coefficient and the intercept for the model
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
```

```python
# Note: we have to transform our min and max values 
# so they are in the format: array([[ 1.17]])
# This is the required format for `model.predict()`

x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
# Calculate the y_min and y_max using model.predict and x_min and x_max
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
# Plot X and y using plt.scatter
# Plot the model fit line using [x_min[0], x_max[0]], [y_min[0], y_max[0]]
plt.scatter(X, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
```
